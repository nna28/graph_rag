```python
import pandas as pd
from pathlib import Path
```

```python
INPUT_FILE = Path(r"C:\Users\ADMIN\Downloads\professors\data\ner\nodes_final.csv")
OUTPUT_FILE = Path(r"C:\Users\ADMIN\Downloads\professors\data\nodes.csv")
```

```python
def extract_final():
    print("Loading:", INPUT_FILE)
    df = pd.read_csv(INPUT_FILE)

    # Check required columns
    required_cols = ["link", "name", "final_label"]
    for c in required_cols:
        if c not in df.columns:
            raise ValueError(f"File thiếu cột bắt buộc: {c}")

    # rename final_label -> type
    df = df.rename(columns={"final_label": "type"})

    # keep only required columns
    out = df[["link", "name", "type"]].copy()

    # drop empty rows
    out = out[out["name"].notna() & out["link"].notna()]
    out = out[out["name"].astype(str).str.len() > 0]

    OUTPUT_FILE.parent.mkdir(parents=True, exist_ok=True)
    out.to_csv(OUTPUT_FILE, index=False)

    print("Saved:", OUTPUT_FILE)
    print("Rows:", len(out))
```

```python
if __name__ == "__main__":
    extract_final()
```
